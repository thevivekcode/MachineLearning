```python
import csv
import time
import numpy as np
from matplotlib import cm
from matplotlib import animation
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xList = np.loadtxt('l./data/q1/linearX.csv', delimiter='\n')
yList = np.loadtxt('./data/q1/linearX.csv', delimiter='\n')
thetaZPlot=[]
thetaOPlot=[]
costPlot=[]

mean = np.mean(xList)
sD= np.std(xList)
for i in range(len(xList)):
    xList[i] = (xList[i]-mean)/sD

    
plt.scatter(xList,yList)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
temp=1
eta = 0.1
x=np.array([np.ones(100),xList])
y=yList.reshape(100,1)
theta = np.zeros(2)
# thetaZPlot.append(0)
# thetaOPlot.append(0)
theta = theta.reshape(2,1)


def costFun(m,theta,x):
    prod=np.dot(theta.T,x)
    diff=y-prod.T
    out=np.square(diff)
    return (1/(2*m))*np.sum(out)


def gradient(y,theta,x):  
    global temp
    prod=np.dot(theta.T,x)
    diff=y-prod.T
    ref1=np.copy(theta[0][0])
    ref2=np.copy(theta[1][0])
    theta[0][0]=ref1+(eta*np.sum(diff))/100
    xlistRef=  xList.reshape(100,1)
    check = np.dot(diff.T,xlistRef)
    theta[1][0]=ref2+(eta*check[0][0])/100
    temp+=1

IC=costFun(100,theta,x)
# costPlot.append(IC)
gradient(y,theta,x)
thetaZPlot.append(theta[0][0])
thetaOPlot.append(theta[1][0])
FC=costFun(100,theta,x)
costPlot.append(FC)

start = time.time()

print(abs(IC-FC))

while(abs(IC - FC) > 1e-15):
    IC=FC
    gradient(y,theta,x)
    thetaZPlot.append(theta[0][0])
    thetaOPlot.append(theta[1][0])
    FC=costFun(100,theta,x)  
    costPlot.append(FC)

print("final theta is")
print(theta)
newY=np.dot(theta.T,x).T
```

```python
 def mesh_support(thetaZSupp,thetaOSupp,m):
    output=[]
    (zTotal,) = thetaZSupp.shape
    (oTotal,) = thetaOSupp.shape
#     print(x.shape)
#     print(zTotal)
    a=0
    b=0

    while(a<zTotal and b<oTotal):
        thetaSupp=np.array([thetaZSupp[a],thetaOSupp[b]])
        thetaSupp=thetaSupp.reshape(1,2)
        prodSupp=np.dot(thetaSupp,x)
        diffSupp=prodSupp.T-y
        outSupp=np.square(diffSupp)
        output.append((1/(2*m))*np.sum(outSupp))
        a+=1
        b+=1
    return output
```

```python
%matplotlib notebook
fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')
# def mesh():   
t0=np.linspace(-1e-6,2,100)
t1=np.linspace(-1,1,100)
X,Y=np.meshgrid(t0,t1)
zSupp=np.asarray(mesh_support(X.flatten(),Y.flatten(),100))
print(zSupp)
Z=zSupp.reshape(X.shape)
axis.plot_surface(X,Y,Z,cmap=cm.Greys, alpha=1)

# mesh()
line, = axis.plot([],[],[],lw=2,color='black',markersize=5)

# def init():
#     line.set_data([],[])
#     line.set_3d_properties([])
#     print("yahan aa bhi rha hai kya")
#     return line,
# print(thetaZPlot)
# print(thetaOPlot)`
# print(costPlot)
length=len(thetaZPlot)
# print(length)

def anime(i):
    print(i)
    xAxis=thetaZPlot
    yAxis=thetaOPlot
#     print(thetaZPlot)
#     print(xAxis)
#     print(yAxis)
#     line, = axis.plot(xAxis,yAxis,costPlot,color='black')
#     print(len(xAxis))
    line.set_data(xAxis[:i],yAxis[:i])
    line.set_3d_properties(costPlot[:i])
    return line,

plottingAnimation = animation.FuncAnimation(fig, anime,frames=length,interval=200,blit='True',repeat='True')
plt.show()
```

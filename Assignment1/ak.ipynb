```python
import numpy as np
from numpy.linalg import matrix_power
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation

def costcomputing(yvector, xvector, theta, m):
    err = (yvector - np.dot(xvector, theta))
    return np.dot(err.T, err)/(2*m)



    
```

```python
def gd():
    filex = open("data/q1/linearX.csv")
    filey = open("data/q1/linearY.csv")
    xlist = []
    ylist = []
    for x,y in zip(filex,filey):
        xlist.append((float)(x.strip()))
        ylist.append((float)(y.strip()))
    xlen = len(xlist)
    ylen = len(ylist)
    xmean = np.mean(np.asarray(xlist))
    xstd = np.std(np.asarray(xlist))
    xlist = [(i-xmean)/xstd for i in xlist]
    xvector = np.array([np.ones(xlen), xlist]).T
    yvector = np.array(ylist).T.reshape(ylen, 1)
    theta = np.array([0, 0]).reshape(2, 1)
    eta = 0.01
    epsilon = 1e-15
    converged = False
    txvector = xvector.T[1].reshape(xlen, 1) # contains the normalized row of x

    jtheta_l = list()
    cost_l = list()
    # ploting to be continued

    while(not converged):
        jtheta = np.zeros(theta.size).reshape(theta.size, 1) # temporary theta initialized to [0, 0]  2 x 1
        err = (yvector - np.dot(xvector, theta))
        jtheta[0] = theta[0] - (eta/(len(ylist))) * np.sum([err[i]*(-xvector[i][0]) for i in range(len(err))])
        jtheta[1] = theta[1] - (eta/(len(ylist))) * np.sum([err[i]*(-xvector[i][1]) for i in range(len(err))])
        for i in theta - jtheta:
            if(abs(i) < epsilon):
                converged = True
            else:
                converged = False
                break

        theta = jtheta.copy()

        # ploting part
        jtheta_l.append(jtheta)
        cost_l.append(np.sum(err**2)/(2*xlen))
        # ploting part end
    return jtheta_l,cost_l,xvector,yvector,xlen
```
